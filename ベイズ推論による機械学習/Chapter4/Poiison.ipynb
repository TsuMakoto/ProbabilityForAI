```julia
# ポアソン分布に従う乱数を生成し、プロット
P(x) = Poi(x, 4)
rand_poi = rejection_method(P, 1000, 0:24)
y = [P(x) for x ∈ rand_poi]
plot(rand_poi, y, st=:bar)
```

```julia
p_observe(xn, λk) = Poi(xn, λk)
p_mix(xn, Sn, Λ) = Π([p(xn, Λ[k])^Sn[k] for k ∈ 1:length(Λ)])
p_before(λk, a, b) = Gam(λk, a, b)
p_s(Sn, PI) = Cat(Sn, PI)
p_p(PI) = Dir(PI, A)
```

```julia
a, b = 2, 2
P_before(x) = p_before(x, a, b);
Λ = rejection_method(P_before, 100, 0:0.01:10);
```

```julia
x,y = Λ, [P_before(λ) for λ ∈ Λ]
plot(x,y, st=:scatter)
```
