```julia
# ポアソン分布に従う乱数を生成し、プロット
P(x) = Poi(x, 4)
rand_poi = rejection_method(P, 1000, 0:24)
y = [P(x) for x ∈ rand_poi]
plot(rand_poi, y, st=:bar)
```

```julia
p_observe(xn, λk) = Poi(xn, λk)
p_mix(xn, Sn, Λ) = Π([p(xn, λk)^snk for snk ∈ Sn])
p_before(λk, a, b) = Gam(λk, a, b)
```
